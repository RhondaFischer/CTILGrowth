## Read parquet file

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
import os
import findspark
findspark.init() 


spark = SparkSession.builder.appName("TwitterParquetReader").getOrCreate()
#df.write.parquet("/tmp/out/people.parquet") 
data_files=[]
data_files += [file for file in os.listdir('parc') if file.endswith(".parquet")]  

print("************************Data_files\n", data_files)

x=0
for file in data_files:
    if x==0:
        parDF1=spark.read.parquet("parc/"+file)
        print('****************************************')
        parDF1.show()
        x+=1
    else:
        parDF2=spark.read.parquet("parc/"+file)
        print('****************************************')
        parDF2.show()
        parDF1 = parDF1.union(parDF2)
        x+=1
parDF1.show()
print("Number of files read",x)
        
```
