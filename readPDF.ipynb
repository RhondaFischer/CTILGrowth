```python
import glob
pdffiles = []
for file in glob.glob("*.pdf"):
    pdffiles.append(file)
```

```python
pdffiles
```

```python
all_text = []
for file in pdffiles:
    print(file)
    pdfFileObj = open(file, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    print(pdfReader.numPages)
    for page in range(1, pdfReader.numPages):
        #print(page)
        pageObj = pdfReader.getPage(page)
        all_text.append(pageObj.extractText())
    pdfFileObj.close()
print(len(all_text))
    
```

```python
'''
import pandas as pd
import json
from datetime import datetime
import matplotlib.pyplot as plt
from collections import Counter
import networkx as nx

import math
import re
import csv
'''
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation
```

```python
# apply Tfidf to vectorize the words
vect = TfidfVectorizer( norm=None, analyzer='word',
                           max_features=700,ngram_range = (1,2),
                          stop_words='english')

# save the vectorized text in "X"
X = vect.fit_transform(all_text)

# create an array of the names
feature_names = np.array(vect.get_feature_names())
```

```python
feature_names
```

```python
# apply NMF to find clusters of words 

number_of_topics = 30
features_per_topic = 7

nmf = NMF(n_components=number_of_topics, solver="mu")
W = nmf.fit_transform(X)
H = nmf.components_

print("Crisis word clusters")
for i, topic in enumerate(H):
 
    print("Topic {}: {}".format(i + 1, ", ".join([str(x) for x in feature_names[topic.argsort()[-features_per_topic:]]])))
```
