## Create a list of links from the main website

```python
# Function for finding links on a website
# input:  website url
# output: list of URLs from those sites

def findlinks(html2):
    while True:
        try:
            r = requests.get(html2, allow_redirects=True,timeout = (2, 5))
            r.raise_for_status()
        except requests.exceptions.RequestException as e: 
            return "" 
        if r.status_code == 200:
            break # Exit loop if status is valid
        else:
            return ""
    sitelinks =[]
    soup = BeautifulSoup(r.text, 'html.parser')
    findAll = soup.find_all("a", href=lambda href: href and href.startswith("h"))
    foundUrls = [link["href"] for link in findAll]
    print("Number of URLs using findlinks:", len(foundUrls))
    return foundUrls
```

## Go to each of the URL's in the main site and find their twitter.com links

```python
def findTwitter(html2):
    foundTweet=[]
    while True:
        try:
            r = requests.get(html2, allow_redirects=True,timeout = (2, 5))
            r.raise_for_status()
        except requests.exceptions.RequestException as e: 
            return "" 
        if r.status_code == 200:
            break # Exit loop if status is valid
        else:
            return ""
    sitelinks =[]
    soup = BeautifulSoup(r.text, 'html.parser')
    findAll = soup.find_all("a", href=lambda href: href and href.startswith("h"))
    foundLink = [link["href"] for link in findAll]
    for link in foundLink: 
        if link.startswith("https://twitter.com"):
            foundTweet.append(link)
    return foundTweet
```

## Instead of returning all links, only return the links that include "twitter.com"

```python
# breakpt is for reducing the amount of links searched for debugging purposes
breakpt = len(haz_links)
site_data = []
print("Number of links on site:" ,len(haz_links))
x=1
for link in haz_links:
    if link.startswith("https://twitter.com"):
        print(link)
        site_data.append(link)
    if x ==breakpt: break
    x+=1
    twitter_links = findTwitter(link)
    while len(twitter_links)>0:
        site_data.append(twitter_links.pop())

print("length of site_data", len(site_data) )   
```
